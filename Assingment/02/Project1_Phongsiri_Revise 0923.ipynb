#2.2 Bigram

```python
#generate bigrams for five sentences
#1st words from unigrams, rest 9 from bigrams
from nltk import FreqDist

cfd_bigram = nltk.ConditionalFreqDist(nltk.bigrams(tokens))
for item in cfd_bigram.items():
    print(item)
```

#2.3 Trigrams

```python
#generate trigrams for five sentences
#1st words from unigrams, 2nd from bigrams, rest 8 from trigrams

trigrams = nltk.trigrams(tokens)
condition_pairs = (((w0, w1), w2) for w0, w1, w2 in trigrams)
cfd_trigram = nltk.ConditionalFreqDist(condition_pairs)

for item in cfd_trigram.items():
    print(item)
```

```python
fdist = FreqDist(cfd_trigram)
fdist.most_common(20)
```

#2.4 Four grams

```python
#generate four grams for five sentences
#1st words from unigrams, 2nd from bigrams, 3rd from trigrams, the rest 4 from four grams

fourgrams = nltk.ngrams(tokens,4)
condition_pairs = (((w0, w1, w2), w3) for w0, w1, w2, w3 in fourgrams)
cfd_fourgram = nltk.ConditionalFreqDist(condition_pairs)

for item in cfd_fourgram.items():
    print(item)
```
