```python
from nltk.tokenize import word_tokenize
```

```python
AI = """Artificial intelligence is a science and technology based on disciplines such as Computer Science, Biology, Psychology, Linguistics, Mathematics, and Engineering. 
A major thrust of AI is in the development of computer functions associated with human intelligence, such as reasoning, learning, and problem solving."""
```

```python
AI_Tokens = word_tokenize(AI)
AI_Tokens
```

```python
len(AI_Tokens)
```

```python
#Find the frequency of the words
from nltk.probability import FreqDist
fdist = FreqDist()
```

```python
for word in AI_Tokens:
  fdist[word.lower()]+=1
fdist
```

```python
#Find the frequency of the word
fdist['functions']
```

```python
#find the unique words in the paragraph
len(fdist)
```

```python
fdist_top10 = fdist.most_common(10)
fdist_top10
```

```python
#Number of the paragraph that given to the new line
from nltk.tokenize import blankline_tokenize
AI_blank = blankline_tokenize(AI)
len(AI_blank)
```

## N-Gram

```python
from nltk.util import bigrams, trigrams, ngrams
```

```python
string = 'The best and most beautiful things in the world cannot be seen or even touched, they must be felt with the heart'
quotes_token = nltk.word_tokenize(string)
quotes_token
```

```python
quotes_bigrams = list(nltk.bigrams(quotes_token))
quotes_bigrams
```

```python
quotes_trigrams = list(nltk.trigrams(quotes_token))
quotes_trigrams
```

```python
#N-gram , N=4
quotes_ngrams = list(nltk.ngrams(quotes_token, 4))
quotes_ngrams
```

## Stemming

```python
from nltk.stem import PorterStemmer
pst = PorterStemmer()
```

```python
pst.stem("having")
```

```python
words_to_stem = ['give','giving','given','gave']
for words in words_to_stem:
  print(words+ ":" + pst.stem(words))
```

```python
from nltk.stem import LancasterStemmer
lst = LancasterStemmer()
for words in words_to_stem:
  print(words+ ":" + lst.stem(words))
  
```

```python
from nltk.stem import SnowballStemmer
sbst = SnowballStemmer("english")
```

```python
for words in words_to_stem:
  print(words+ ":" + sbst.stem(words))
```

## POS : Part of Speech
this part show the type of the words in the sentence, such as verb, noun, pronoun, and determination.

```python
 nltk.download('averaged_perceptron_tagger')
 sent = "Timothy is a natural when it comes to drawing"
 sent_tokens = word_tokenize(sent)
```

```python
for token in sent_tokens:
  print(nltk.pos_tag([token]))
```

```python
sent2 = "John is eating a delicious cake"
sent2_tokens = word_tokenize(sent2)
for token in sent2_tokens:
  print(nltk.pos_tag([token]))
```

## NER = Named Entity Recognition

```python
from nltk import ne_chunk
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
NE_sent = "The US President stays in the White House"
```

```python
NE_tokens = word_tokenize(NE_sent)
NE_tags = nltk.pos_tag(NE_tokens)
```

```python
NE_NER = ne_chunk(NE_tags)
print(NE_NER)
```
