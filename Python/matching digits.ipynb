```python
import re
p = re.compile(r'\d+')
r = p.findall('12 drummers drumming, 11 piper piping, 10 lords a leaving')
print(r)
print(re.match(r'From\s+', "From this point forward"))
```

```python
p = re.compile('ab*',re.IGNORECASE)
print(p)
```

```python
p = re.compile('(a(b)c)d')
m = p.match('abcd')
print(m.group(0))
print(m.group(1))
print(m.group(2))
print(m.group(2,1,0))
```

```python
p = re.compile(r'\b(\w+)\s+\1\b')
p.search('Paris in the the spring').group()
```
