```python
text = "Baker Betty Lou bought some butter. But, it made her batter bitter. So, Baker Betty Lou bought some better butter to make her bitter batter better."


# find the most pair of alphabet that appeals together

def find_most_common_pairs(text):
    my_dict = dict()
    for i, v1 in enumerate(text[:-1]):
        j = i + 1
        v2 = text[j]
        if (f'{v1}{v2}') in my_dict:
            my_dict[f'{v1}{v2}'] += 1
        else:
            my_dict[f'{v1}{v2}'] = 1
    my_dict = {k: v for k, v in sorted(my_dict.items(), key=lambda x: x[1], reverse=True) if len(k.strip()) == 2}
    return my_dict


find_most_common_pairs(text)
```
